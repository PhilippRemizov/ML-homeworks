```python
from sklearn.model_selection import train_test_split
```

```python
hous_df.shape
```

```python
# x = hous_df.drop(['MEDV','ZN','B'],axis=1) лучше не стало
x = hous_df.drop('MEDV',axis=1)
y = hous_df['MEDV']
```

```python
X_train, X_test, y_train, y_test = train_test_split(x,y, test_size=0.22, random_state=55)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model_lr = LinearRegression()
```

```python
model_lr.fit(X_train,y_train)
```

```python
pred = model_lr.predict(X_test)
```

```python
pred
```

```python
y_test
```

```python
from sklearn.metrics import mean_squared_error
```

```python
print('Ошибка MSE=',mean_squared_error(y_test,pred,squared=True))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(20,9))
plt.plot(y_test.values, "blue", linewidth=5, label = 'Реальные значения')
plt.plot(pred, "red", linewidth=5, label = 'Наш первый ML')
plt.legend(loc='best')
plt.grid(True)
plt.show()
```
